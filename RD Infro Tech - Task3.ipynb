```python
#RD Infro Tech - Task3
#STEP-1
#import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
#Load Dataset using pandas
data_frame=pd.read_csv("Titanic-Dataset.csv") 
data_frame.info()
#STEP-2
# Data Cleaning
# checking for missing values
missing_values=data_frame.isnull().sum()
print(missing_values)
# Handling missing values Age,Embarked,Cabin
Age=data_frame['Age'].fillna(data_frame['Age'].mean())
print(Age)
Embarked=data_frame['Embarked'].fillna(data_frame['Embarked'].mode()[0])
print(Embarked)
data_frame.drop('Cabin',axis=1,inplace=True)
# Rechecking missing values Again for if any missing data remains
missing_values=data_frame.isnull().sum()
print(missing_values)
# Checking Duplicated data
Duplicates=data_frame.duplicated().sum()
print("Duplicates =", Duplicates)
#STEP-3
# Data Preprocessing
#Encode categorical Data
df=pd.get_dummies(data_frame, drop_first=True)
# Feature Scaling
scalar = StandardScaler()
scaled_data = scalar.fit_transform(df)
```
